```python
text = """
동해물과 백두산이 마르고 닳도록
하느님이 보우하사 우리나라 만세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
남산 위에 저 소나무 철갑을 두른 듯
바람 서리 불변함은 우리 기상일세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
가을 하늘 공활한데 높고 구름 없이
밝은 달은 우리 가슴 일편단심일세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
이 기상과 이 맘으로 충성을 다하여
괴로우나 즐거우나 나라 사랑하세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세"""
text = text * 10
print(text)
```

```python

# 문자 집합 생성
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(chars)
print(vocab_size)

# 문자 ↔ 인덱스 매핑
char2idx = {ch: i for i, ch in enumerate(chars)}
idx2char = {i: ch for i, ch in enumerate(chars)}

print(char2idx)
print(idx2char)
```

```python
print(len(text))
data = [char2idx[ch] for ch in text]
print(data)
```

```python
import torch
import numpy as np

seq_len = 30
X = []
Y = []
for i in range(len(data) - seq_len):
    X.append(data[i:i+seq_len])
    Y.append(data[i+seq_len])

X = torch.tensor(X, dtype=torch.long)
Y = torch.tensor(Y, dtype=torch.long)

print(X.shape)
print(Y.shape)

print(X[0])
print(Y[0])

sample_0_x = [idx2char[i.item()] for i in X[0]] #텐서 숫자 변환 후 사전 조회
print(sample_0_x)

sample_0_y = [idx2char[Y[0].item()]] #텐서 숫자 변환 후 사전 조회
print(sample_0_y)
```

```python
import torch.nn as nn

class CharLSTM(nn.Module):
    def __init__(self, vocab_size, hidden_size, num_layers=1):
        super().__init__()
        #Embedding Layer
        #(vocab_size, hidden_size) = (84, 128)
        # 입력: (batch, seq_len) -> 출력( batch, seq_len, 128)
        self.embed = nn.Embedding(vocab_size, hidden_size)

        #batch_first를 True로 하면
        #입력 shape: (batch, seq_len, input_size)가 되어 조금 더 직관적이긴 합니다.
        #이런 옵셥이 있다는 것을 알면 됩니다.
        #출력 sahpe: (batch, seq_len, hidden_size)가 됩니다.
        #실제 입력: (2820, 30, 128)
        #실제 출력: (2820, 30, 128)
        self.lstm = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)

        #fully connected llaeyr
        #입력 (batch, 128) - > 출력(batch, 84)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x, hidden=None): #hidden=None, 초기 은닉상태 h_0, c_0를 0으로
        # x: (batch, seq_len) = (2820, 30)
        x = self.embed(x) #(2820, 30, 128)
        #2820개의 시퀀스, 각 시퀀스는 30개 문자, 각 문자는 128차원 벡터

        # out: 모든 시점의 은닉 상태 → (B, T, 128) = (2820, 30, 128)
        # hidden: (h_n, c_n) 튜플 ( (1,2820,128), (1,2820,128) )
        #   h_n: 마지막 시점 은닉 상태 → (num_layers, B, 128) = (1, 2820, 128)
        #   c_n: 마지막 시점 셀 상태   → (num_layers, B, 128) = (1, 2820, 128)

        out, hidden = self.lstm(x, hidden)

        # 마지막 시점만 사용: out[:, -1] → (B, 128) = (2820, 128)
        # FC 통과: (2820, 84)
        out = self.fc(out[:, -1])
        return out, hidden

```

```python
import torch
import torch.nn as nn


# 입력 시퀀스 (batch 없이)
test_x = torch.tensor([char2idx[ch] for ch in ['동', '해', '물', '과']], dtype=torch.long)
print("입력 인덱스:", test_x)
print("입력 shape:", test_x.shape)

# 임베딩 정의 (3차원)
embedding = nn.Embedding(num_embeddings=len(char2idx), embedding_dim=3)

# 임베딩 실행
out = embedding(test_x)

# 올바른 문자-벡터 출력
print("\n임베딩 결과 shape:", out.shape)
for idx, vec in zip(test_x, out):
    print(f"{idx2char[idx.item()]} → {vec.tolist()}")
```

```python
print(f"vocab size: {vocab_size}")
model = CharLSTM(vocab_size=vocab_size, hidden_size=128)
loss_fn = nn.CrossEntropyLoss() #84class
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
def generate_text(model, start_text, length):
    # 모델을 추론 모드로 설정
    model.eval()

    # 시작 문자열을 정수 시퀀스로 변환 (배치 차원 추가)
    input_seq = torch.tensor([char2idx[ch] for ch in start_text], dtype=torch.long).unsqueeze(0)

    # 생성된 텍스트를 저장할 리스트
    generated = list(start_text)

    # LSTM 초기 은닉 상태
    hidden = None

    # length만큼 반복해서 새로운 문자 생성
    for _ in range(length):
        # 모델 통과 → 다음 글자 확률 분포 얻기
        output, hidden = model(input_seq, hidden)

        # softmax로 확률 분포 계산
        probs = nn.functional.softmax(output, dim=-1).squeeze()

        # 확률 분포에 따라 랜덤 샘플링
        next_idx = torch.argmax(probs).item()
        #next_idx = torch.multinomial(probs, 1).item()
        next_char = idx2char[next_idx]
        generated.append(next_char)

        # 다음 입력 시퀀스 갱신 (맨 앞 제거 + 새 문자 추가)
        input_seq = torch.cat([
            input_seq[:, 1:],
            torch.tensor([[next_idx]])
        ], dim=1)

    return ''.join(generated)
```

```python
import matplotlib.pyplot as plt

n_epochs = 200
loss_history = []  # 에폭별 손실 저장 리스트

for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()

    output, _ = model(X)  # output: (batch, vocab_size)
    loss = loss_fn(output, Y)  # 정답 Y와 예측 output을 비교
    loss.backward()
    optimizer.step()

    # 손실 기록
    loss_history.append(loss.item())

    # 주기적으로 출력 (손실 + 샘플 문장 생성)
    if epoch % 20 == 0:
        print(f"Epoch {epoch:3d} | Loss: {loss.item():.4f}")

        # 가을 하늘로 시작하는 샘플 생성
        sample = generate_text(model, start_text="가을 하늘 ", length=50)
        print("샘플 생성:", sample)
        print("-" * 80)

# 훈련 후 손실 그래프 시각화
plt.plot(range(n_epochs), loss_history)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.grid(True)
plt.show()
```
